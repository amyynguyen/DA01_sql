# 1. Clustering Gaussian Blobs using $k$-means

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

# Generate 5 Gaussian blobs in 10 dimensions
X, y_true = make_blobs(
    n_samples=1000,
    centers=5,
    n_features=10,
    cluster_std=1.5,
    random_state=1)        # reproducibility
X = StandardScaler().fit_transform(X)

print(type(X),X.shape)
print(type(y_true),y_true.shape)
```

```python
# your code here
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix
from scipy.optimize import linear_sum_assignment
import matplotlib.pyplot as plt
# K-mean with k = 5
kmeans = KMeans(n_clusters=5, random_state=1, n_init=10)
y_pred = kmeans.fit_predict(X)
print("K-means Inertia (k=5):", kmeans.inertia_)

# PCA visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
centers_pca = pca.transform(kmeans.cluster_centers_)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_pred, cmap='viridis')
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='black', s=200, alpha=0.5)
plt.title("K-means Clusters (PCA 2D)")
plt.show()

# Confusion matrix (align labels)
cm = confusion_matrix(y_true, y_pred)
row_ind, col_ind = linear_sum_assignment(cm.max() - cm)
cm_aligned = cm[np.ix_(row_ind, col_ind)]
print("Confusion Matrix (k=5):", cm_aligned)
```
